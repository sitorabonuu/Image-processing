### <div style="direction:ltr;text-align:left;font-family:Arial, Tahoma">Displaying in the Notebook with matplotlib</div>

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
In fact, the rows and columns of the matrix correspond to the height and width of the image, respectively.<br/>
The image colors appear inverted because the channel order in OpenCV is BGR, unlike most other libraries which use RGB.</div>

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Changing the color channel order (BGR → RGB)</div>


<div style="direction:rtl;text-align:right;font-family:Tahoma">
Method 1: Using NumPy</div>

```python
#method1
plt.imshow(img[:,:,::-1])
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
Method 2: Using OpenCV</div>

```python
#method2
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```
