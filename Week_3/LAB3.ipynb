```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("C:/Users/sudob/OneDrive/Desktop/Week_3/image.png")

print("Image shape:", image.shape)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
```

```python
tx = 100
ty = 50

translation_matrix = np.float32([[1,0,tx],
                                 [0,1,ty]])

translated_image = cv2.warpAffine(image,
                                  translation_matrix,
                                  (image.shape[1], image.shape[0]))

plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("Translated Image")
plt.axis("off")
```

```python
tx = -80
ty = -40

translation_matrix = np.float32([[1,0,tx],
                                 [0,1,ty]])

translated_image = cv2.warpAffine(image,
                                  translation_matrix,
                                  (image.shape[1], image.shape[0]))

plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("Negative Translation")
plt.axis("off")
```

```python
(h, w) = image.shape[:2]
center = (w//2, h//2)

rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1)

rotated_image = cv2.warpAffine(image, rotation_matrix, (w,h))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title("Rotation 45°")
plt.axis("off")
```

```python
rotation_matrix = cv2.getRotationMatrix2D(center, 90, 1)

rotated_image = cv2.warpAffine(image, rotation_matrix, (w,h))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title("Rotation 90°")
plt.axis("off")
```

```python
factor = 2

scaling_matrix = cv2.getRotationMatrix2D((0,0),0,factor)

scaled_image = cv2.warpAffine(image,
                              scaling_matrix,
                              (w*factor, h*factor))

plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
plt.title("Scaled Image (2x)")
plt.axis("off")

print("Scaled shape:", scaled_image.shape)
```

```python
factor = 0.5

scaling_matrix = cv2.getRotationMatrix2D((0,0),0,factor)

scaled_image = cv2.warpAffine(image,
                              scaling_matrix,
                              (int(w*factor), int(h*factor)))

plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
plt.title("Scaled Image (0.5x)")
plt.axis("off")
```

```python
m = 0.67

shearing_matrix = np.float32([[1, m, 0],
                              [0, 1, 0]])

sheared_image = cv2.warpAffine(image,
                               shearing_matrix,
                               (w,h))

plt.imshow(cv2.cvtColor(sheared_image, cv2.COLOR_BGR2RGB))
plt.title("Sheared Image")
plt.axis("off")
```

```python
cv2.imwrite("translated.png", translated_image)
cv2.imwrite("rotated.png", rotated_image)
cv2.imwrite("scaled.png", scaled_image)
cv2.imwrite("sheared.png", sheared_image)
```
