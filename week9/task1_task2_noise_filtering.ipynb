# Image Noise and Filtering - Task 1 & Task 2

This notebook includes:
- Task 1: Gaussian noise + Median filter
- Task 2: Gaussian noise (variance 0.0009) + Mean and Geometric Mean filters


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show(image, title):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Load image (update path if needed)
img = cv2.imread('input.jpg', 0)

show(img, 'Original Image')
```

```python
# Gaussian Noise Function
def add_gaussian_noise(image, mean=0, var=0.01):
    sigma = var ** 0.5
    noise = np.random.normal(mean, sigma, image.shape)
    noisy = image + noise * 255
    return np.clip(noisy, 0, 255).astype(np.uint8)
```

```python
# Task 1: Gaussian noise + Median filter
noisy1 = add_gaussian_noise(img, 0, 0.01)
median_filtered = cv2.medianBlur(noisy1, 3)

show(noisy1, 'Gaussian Noise (var=0.01)')
show(median_filtered, 'Median Filter 3x3')
```

```python
# Mean filter
def mean_filter(image, ksize):
    return cv2.blur(image, (ksize, ksize))

# Geometric mean filter (approximation)
def geometric_mean_filter(image, ksize):
    image = image.astype(float) + 1e-5
    log_image = np.log(image)
    kernel = np.ones((ksize, ksize)) / (ksize * ksize)
    mean_log = cv2.filter2D(log_image, -1, kernel)
    geo = np.exp(mean_log)
    return np.clip(geo, 0, 255).astype(np.uint8)
```

```python
# Task 2: Gaussian noise (var=0.0009)
noisy2 = add_gaussian_noise(img, 0, 0.0009)

mean3 = mean_filter(noisy2, 3)
mean5 = mean_filter(noisy2, 5)

geo3 = geometric_mean_filter(noisy2, 3)
geo5 = geometric_mean_filter(noisy2, 5)

show(noisy2, 'Gaussian Noise (var=0.0009)')
show(mean3, 'Mean Filter 3x3')
show(mean5, 'Mean Filter 5x5')
show(geo3, 'Geometric Mean Filter 3x3')
show(geo5, 'Geometric Mean Filter 5x5')
```
