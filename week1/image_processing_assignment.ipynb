# Image Reading and Processing Assignment
This notebook demonstrates how to read and process images using PIL, NumPy, Matplotlib, scikit-image, and OpenCV.

##  Using PIL (Pillow)

```python
from PIL import Image

img = Image.open("C:/Users/sudob/OneDrive/Desktop/week1/images (1).jpg")

print('Image type:', type(img))
print('Image size:', img.size)
print('Image mode:', img.mode)

img
```

## Convert PIL Image to NumPy Array

```python
import numpy as np

img_array = np.array(img)

print('Array shape:', img_array.shape)
print('Data type:', img_array.dtype)
```

##  Read Image Using Matplotlib

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img2 = mpimg.imread("C:/Users/sudob/OneDrive/Desktop/week1/bear.jpg")

print('Image shape:', img2.shape)

plt.imshow(img2)
plt.axis('off')
plt.show()
```

### Concept Answer
Matplotlib treats images as NumPy arrays because images consist of pixel values arranged in a grid. NumPy arrays efficiently store and manipulate numerical data, which makes them suitable for image processing.

##  Using scikit-image

```python
from skimage import io, img_as_float

img3 = io.imread("C:/Users/sudob/OneDrive/Desktop/week1/bear.jpg")
img_float = img_as_float(img3)

print('Min pixel value:', img_float.min())
print('Max pixel value:', img_float.max())

plt.imshow(img_float)
plt.axis('off')
plt.show()
```

### Explanation
`astype(float)` only changes the data type but does not scale pixel values. `img_as_float` converts the image and scales pixel values to the range 0–1, which is more appropriate for image processing tasks.

##  Using OpenCV

```python
import cv2

gray = cv2.imread("C:/Users/sudob/OneDrive/Desktop/week1/images (1).jpg", 0)
color = cv2.imread("C:/Users/sudob/OneDrive/Desktop/week1/images (1).jpg")

rgb = cv2.cvtColor(color, cv2.COLOR_BGR2RGB)

plt.imshow(rgb)
plt.axis('off')
plt.show()
```

##  Edge Detection

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```
