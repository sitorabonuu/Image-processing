```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("C:/Users/sudob/OneDrive/Desktop/week12/rice.png")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')
plt.show()
```

```python
_, thresh = cv2.threshold(gray, 119, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()
```

```python
import numpy as np
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

plt.imshow(opening, cmap='gray')
plt.title("After Noise Removal")
plt.axis('off')
plt.show()
```

```python
contours, _ = cv2.findContours(opening, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
count = len(contours)
print("Number of rice grains:", count)
```

```python
output = image.copy()
cv2.drawContours(output, contours, -1, (0,255,0), 2)

plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.title(f"Detected Grains: {count}")
plt.axis('off')
plt.show()
```
