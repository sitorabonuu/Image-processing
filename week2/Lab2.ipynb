```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("C:/Users/sudob/OneDrive/Desktop/week2/tiger.jpg")

image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_RGB)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
sub_30 = cv2.subtract(image_RGB, 30)
sub_100 = cv2.subtract(image_RGB, 100)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(image_RGB)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(sub_30)
plt.title("Subtract 30")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(sub_100)
plt.title("Subtract 100")
plt.axis("off")

plt.show()
```

```python
R, G, B = cv2.split(image_RGB)

R_sub = cv2.subtract(R, 100)

red_reduced = cv2.merge((R_sub, G, B))

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(red_reduced)
plt.title("Red Channel Reduced")
plt.axis("off")

plt.show()
```

```python
add_30 = cv2.add(image_RGB, 30)
add_100 = cv2.add(image_RGB, 100)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(image_RGB)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(add_30)
plt.title("Add 30")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(add_100)
plt.title("Add 100")
plt.axis("off")

plt.show()
```

```python
R_add = cv2.add(R, 100)

red_increased = cv2.merge((R_add, G, B))

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(red_increased)
plt.title("Red Channel Increased")
plt.axis("off")

plt.show()
```

```python
multiply_half = cv2.multiply(image_RGB, 0.5)
multiply_two = cv2.multiply(image_RGB, 2)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(image_RGB)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(multiply_half)
plt.title("Multiply 0.5")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(multiply_two)
plt.title("Multiply 2")
plt.axis("off")

plt.show()
```

```python
divide_two = cv2.divide(image_RGB, 2)
divide_three = cv2.divide(image_RGB, 3)
divide_half = cv2.divide(image_RGB, 0.5)

plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(image_RGB)
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(divide_two)
plt.title("Divide 2")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(divide_three)
plt.title("Divide 3")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(divide_half)
plt.title("Divide 0.5")
plt.axis("off")

plt.show()
```
