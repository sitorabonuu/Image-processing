```python
def box_filt(n):

    if n <= 0 or n % 2 == 0:
        raise ValueError("Window size must be positive odd integer")

    kernel = np.ones((n,n),np.float32)/(n*n)
    return kernel
```

```python
kernel = box_filt(5)

box_filtered = cv2.filter2D(image,-1,kernel)

plt.imshow(box_filtered,cmap='gray')
plt.title("Box Filtered Image")
plt.axis("off")
plt.show()
```

```python
blurred = cv2.GaussianBlur(image,(9,9),10)

unsharp = cv2.addWeighted(image,1.5,blurred,-0.5,0)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(blurred,cmap='gray')
plt.title("Blurred Image")

plt.subplot(1,2,2)
plt.imshow(unsharp,cmap='gray')
plt.title("Unsharp Masking")

plt.show()
```

```python
def apply_filters(image_input, box, filt_size):

    pad_size = int(np.ceil(filt_size/2))

    image_padded = np.pad(image_input,
                          pad_width=((pad_size,pad_size),(pad_size,pad_size)),
                          mode='symmetric')

    image_box = copy.deepcopy(image_input)

    row, column = image_input.shape

    for i in range(row):
        for j in range(column):

            patch_curr = image_padded[i:i+filt_size, j:j+filt_size]

            results_box = box * patch_curr

            image_box[i,j] = np.sum(results_box)

    return image_box
```

```python
image_input = cv2.imread("C:/Users/sudob/OneDrive/Desktop/week6/images.jpg",0).astype('float32')/255

filt_size = 5
k = 2

box = box_filt(filt_size)

image_blur = apply_filters(image_input, box, filt_size)

image_diff = image_input - image_blur

image_sharp = image_input + k*image_diff

plt.imshow(image_sharp,cmap='gray')
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```

```python
filt_sizes = [3,5,7,9,11,15]
k_values = [-2,-1,-0.5,0,0.5,1,2,3,5,9]
```
